# Data Cleaning

```python
# Dropping Unnecessary columns
df.drop(['id','date'], inplace=True, axis=1)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Function to remove ambiguous values
def remove_ambiguous_values(df, column, ambiguous_value):
    unique_values_before = df[column].unique()
    print(f"Unique values in '{column}' before removal: {unique_values_before}")
    print(f"Removing rows where '{column}' has the ambiguous value: {ambiguous_value}")
    df_cleaned = df[df[column] != ambiguous_value]
    unique_values_after = df_cleaned[column].unique()
    print(f"Unique values in '{column}' after removal: {unique_values_after}")
    return df_cleaned
df_cleaned = remove_ambiguous_values(df, 'bedrooms', 0)
```

```python
# Rounding bathrooms as integer
df.loc[df_cleaned.index, 'bathrooms_int'] = df_cleaned['bathrooms'].round().astype(int)
df['bathrooms_int'] = df['bathrooms_int'].fillna(df['bathrooms'].round().astype(int))
```

```python
# Function to check and  remove outliers
def check_and_remove_outliers_zscore(df, column, threshold=3, remove=False):
    data = df[column]
    z_scores = np.abs((data - np.mean(data)) / np.std(data))
    outliers = df[z_scores > threshold]

    print(f"Number of outliers in '{column}': {len(outliers)}")

    if remove:
        # Removing the outliers
        df = df[z_scores <= threshold]
        print(f"Shape of DataFrame after removing outliers: {df.shape}")
        return df
    else:
        return outliers

df_cleaned = check_and_remove_outliers_zscore(df, 'price', remove=True)
print(df_cleaned.shape)
```
